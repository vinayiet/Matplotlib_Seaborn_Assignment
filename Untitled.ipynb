```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(50)

# Number of samples
num_samples = 100

# Generate Feature1
Feature1 = np.random.normal(loc=50, scale=10, size=num_samples)

# Generate Feature2 correlated with Feature1
Feature2 = Feature1 * 0.5 + np.random.normal(loc=0, scale=5, size=num_samples)

# Generate Feature3 correlated with both Feature1 and Feature2
Feature3 = Feature1 * 0.3 + Feature2 * 0.7 + np.random.normal(loc=0, scale=5, size=num_samples)

# Create DataFrame
df = pd.DataFrame({
    'Feature1': Feature1,
    'Feature2': Feature2,
    'Feature3': Feature3
})

# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)

# Add title and labels
plt.title('Correlation Matrix Heatmap')
plt.xlabel('Features')
plt.ylabel('Features')

# Show the plot
plt.show()
```

```python
# Using the sales data, generate a heatmap to visualize the variation in sales across
# different months and days.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df[['Day', 'Sales']].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="sin(x)", line_width=2)
output_notebook()
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Generate random x and y values
np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

# Generate random sizes and colors
sizes = np.random.rand(100) * 50 + 10  # Sizes between 10 and 60
colors = np.random.choice(["red", "blue", "green", "orange", "purple"], 100)

# Create a DataFrame
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with different sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=data)

# Display the plot in the notebook
output_notebook()
show(p)

```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Display the plot in the notebook
output_notebook()
show(p)
```
